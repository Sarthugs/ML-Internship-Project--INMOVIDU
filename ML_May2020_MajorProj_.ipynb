1) Logistic Regression

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('project.csv', index_col=0)
```

```python
data
```

```python
data=pd.DataFrame(data,columns=[ 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'male', 'Q',
       'S','Survived',])
```

```python
data
```

```python
x=data.iloc[:,:-1].values
```

```python
print(x.shape)
```

```python
x
```

```python
y=data.iloc[:,-1].values
```

```python
print(y.shape)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
pd.DataFrame([xtrain.shape , xtest.shape  , ytrain.shape  ,  ytest.shape],index=['xtrain','xtest','ytrain','ytest'],
             columns=['row','columns'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression 
LR=LogisticRegression()
```

```python
LR.fit(xtrain,ytrain)
```

```python
ypred=LR.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cmat=confusion_matrix(ytest,ypred)
print(cmat)
```

```python
TP=cmat[1,1]
TN=cmat[0,0]
```

```python
FP=cmat[0,1]
FN=cmat[1,0]
pd.DataFrame([TP,TN,FP,FN],index=['TP','TN','FP','FN'])
```

```python
from sklearn import metrics as met
```

```python
accuracy=met.accuracy_score(ytest,ypred)
print("classification accuracy:",accuracy*100)
```

```python
print("classification error is :",100*(1-accuracy))
```

```python
TPR=100*met.recall_score(ytest,ypred)
TNR=100*TN/(TN+FP)
FPR=100-TNR
FNR=100-TPR
```

```python
pd.DataFrame([TPR,TNR,FPR,FNR],index=['TPR/SENSITIVITY','TNR/SPECIFICITY','FPR','FNR'])
```

```python
from sklearn.metrics import precision_score,f1_score,roc_auc_score
precision=precision_score(ytest,ypred)
f1=f1_score(ytest,ypred)
roc_auc=roc_auc_score(ytest,ypred)
```

```python
pd.DataFrame([precision,f1,roc_auc],index=['precision','F1_score','roc_auc'])
```

The accuracy of the model is 84.3575%.

```python
error=[]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i in range(1,40,2):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    yp=knn.predict(xtest)
    error.append(np.mean(yp!=ytest))
plt.figure(figsize=(10,10))
plt.plot(range(1,40,2),error,color='b',linestyle='dashed',marker='*',markerfacecolor='r',markersize='10')
plt.xlabel('k value')
plt.ylabel('mean error')
plt.grid(True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics as met
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc=[]
```

```python
for i in range(1,95,1):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=i/100,random_state=5)
    scaler=StandardScaler()
    xtrain=scaler.fit_transform(xtrain)
    xtest=scaler.transform(xtest)
    knn=KNeighborsClassifier(n_neighbors=21)
    knn.fit(xtrain,ytrain)
    ypred=knn.predict(xtest)
    accuracy=met.accuracy_score(ytest,ypred)
    acc.append(accuracy)
plt.figure(figsize=(10,10))
plt.plot(range(1,95,1),acc,color='g',linestyle='dashed',marker='.',markerfacecolor='r',markersize='10')
plt.xlabel('test size')
plt.ylabel('accuracy')
plt.grid(True)
```

```python
print(acc.index(max(acc)),'-',max(acc))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics as met
from sklearn.tree import DecisionTreeClassifier
acc=[]
```

```python
for i in range(1,100,1):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=i/100,random_state=5)
    scaler=StandardScaler()
    xtrain=scaler.fit_transform(xtrain)
    xtest=scaler.transform(xtest)
    dt=DecisionTreeClassifier()
    dt.fit(xtrain,ytrain)
    ypred=dt.predict(xtest)
    accuracy=met.accuracy_score(ytest,ypred)
    acc.append(accuracy)
plt.figure(figsize=(10,10))
plt.plot(range(1,100,1),acc,color='g',linestyle='dashed',marker='.',markerfacecolor='r',markersize='10')
plt.xlabel('test size(%)')
plt.ylabel('accuracy')
plt.grid(True)
print(acc.index(max(acc)),'-',max(acc))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics as met
from sklearn.ensemble import RandomForestClassifier
acc=[]
```

```python
for i in range(1,100,1):
    xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=i/100,random_state=6)
    scaler=StandardScaler()
    xtrain=scaler.fit_transform(xtrain)
    xtest=scaler.transform(xtest)
    rf=RandomForestClassifier(n_estimators=50,n_jobs=10,random_state=6)
    rf.fit(xtrain,ytrain)
    ypred=rf.predict(xtest)
    accuracy=met.accuracy_score(ytest,ypred)
    acc.append(accuracy)
plt.figure(figsize=(10,10))
plt.plot(range(1,100,1),acc,color='g',linestyle='dashed',marker='.',markerfacecolor='r',markersize='10')
plt.xlabel('test size')
plt.ylabel('accuracy')
plt.grid(True)
print(acc.index(max(acc)),'-',max(acc))
```

From the graph, accuracy can have maximum of 84% and from the calculations 83.7988% with the exception for test size of 1% for which accuracy reaches 100% since the number of samples get reduced. Hence we cannot consider 1% test size for building the mode.

RESULT :

From the above classification models, KNN classifier has the highest accuracy on randomly selected samples for a given test split.
